```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.optimize import minimize

from mpl_toolkits import mplot3d
```

```python
#visualize first, then enter initial guesses?
#I.E., SET UP THE 3D PLOTTING.


f = lambda x1,x2: -(1-x1**2-x2**2)/(np.sqrt(np.abs(1-x1**2-x2**2)))

x = np.linspace(-20,20,1000)
y = np.linspace(-20,20,1000)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,100,cmap='gray')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0,30)
plt.show()

#guess x1,x2 given visualization of function --> this option is at least available
#for functions with two parameters.
one = input('initial guess x:')
two = input('initial guess y:')
x0 = (float(one),float(two))
```

```python
#defining function now with compatible syntax (x[0], x[1] are "one" and "two" entries
#from above, respectively)

f = lambda x: -(1-x[0]**2-x[1]**2)/(np.sqrt(np.abs(1-x[0]**2-x[1]**2)))

#execute minimize, x0 are initial parameter guesses from above
result = minimize(f,x0,method='SLSQP')
```
