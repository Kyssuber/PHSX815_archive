```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.integrate import quadrature
#import sys

#sys.path.append(".")

#for class, input must have form -func lambda x: function

#output will give numerical result (Gaussian quadrature approximation to integral) and the error (difference
#between the last two estimates of the integral)
def GausQuad(func,a,b):
    print('val GausQuad:', quadrature(func,a,b)[0])
    print('error GausQuad:',quadrature(func,a,b)[1])

    
#main function
#if __name__ == "__main__":
    
```

```python
#calculates integral from a to b for some function numerically using trapezoid rule
#be sure function >= 0.

#the following will all be command line inputs once code is finalized...currently in testing phase. :-)
func = np.sin(4*x)**2
#x array for plotting func
x = np.linspace(0,3,200)
a = 0  # start bound
b = 3  # stop bound
N = 30 # num subintervals
f = lambda x: np.sin(4*x)**2 
#x array for trapezoid rule
X = np.linspace(a,b,N+1)


def TrapRule():
    
    #function
    y = f(x)
    plt.figure(figsize=(8,6))
    plt.plot(x,y)
    
    
    #the trapezoids!
    for i in range(N):
        xs = [X[i],X[i],X[i+1],X[i+1]]
        ys = [0,f(X[i]),f(X[i+1]),0]
        plt.fill(xs,ys,'b',edgecolor='b',alpha=0.2)
    
    plt.title('Trapezoid Rule, N = {}'.format(N),fontsize=16)
    plt.xlabel(r'$x$',fontsize=16)
    plt.ylabel(r'$f(x)$',fontsize=16)
    plt.show()
```

```python
TrapRule()
```
